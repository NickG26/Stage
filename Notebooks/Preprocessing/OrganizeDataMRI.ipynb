```python
directory = 'Images/MRIscans/kaggle_3m'
```

```python
content_dir = os.listdir(directory)
print(len(content_dir))
content_dir[:5]
```

```python
dataset = {
    'x' : [],
    'y' : []
}
```

```python
for sub_dir in content_dir:
    dir_patient = f'{directory}/{sub_dir}'
    if os.path.isdir(dir_patient):
        nr_images = int(len(os.listdir(dir_patient)) / 2)

        for i in range(1, nr_images + 1):
            dataset['x'].append(f'{dir_patient}/{sub_dir}_{i}.tif')
            dataset['y'].append(f'{dir_patient}/{sub_dir}_{i}_mask.tif')
```
