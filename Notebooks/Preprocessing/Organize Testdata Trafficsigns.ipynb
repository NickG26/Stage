```python
import pandas as pd
import os
import shutil
```

```python
test_ds = pd.read_csv('Images/traffic-signs/Test.csv')
test_ds.head()
```

```python
for i in range(43):
    directory = f'Images/traffic-signs/Test/{i}'
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
amount_images = test_ds['Path'].count()
image_path = ""

for i in range(amount_images):
    image_path = f"Images/traffic-signs/{test_ds['Path'][i]}"
    if os.path.exists(image_path):
        target_path = f"Images/traffic-signs/Test/{test_ds['ClassId'][i]}"
        shutil.move(image_path, target_path)
#         print(image_path)
#         print(target_path)
#     break;
    
```
